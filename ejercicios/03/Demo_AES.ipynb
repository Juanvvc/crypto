# Demostración de cifrado simétrico: AES

Esta demostración está pensada para seguirse durante una clase interactiva.

```python
!pip install pycryptodome

import Crypto.Random
import Crypto.Cipher.AES
from Crypto.Util.Padding import pad, unpad
from base64 import b64encode, b64decode
```

Creación de una clave aleatoria de 32 bytes

```python
key = Crypto.Random.get_random_bytes(32)
```

```python
print(f'Clave como bytes: {str(key)}')
```

```python
print(f'Clave como número entero: {int.from_bytes(key)}')
```

```python
print(f'Clave como codificada en Base64: {b64encode(key)}')
```

```python
print(f'Logitud de la clave en bytes: {len(key)}')
```

```python
cipher = Crypto.Cipher.AES.new(key, Crypto.Cipher.AES.MODE_ECB)
```

```python
cipher.encrypt('hola')
# error: no podemos cifrar cadenas, solo bytes
```

```python
cipher.encrypt(b'hola')
# error: AES tiene una logitud de bloque de 128bits, 16 bytes
```

```python
print(pad(b'hola', 16))
```

```python
print(cipher.encrypt(pad(b'hola', 16)))
```

```python
print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))
```

# Emisor (modo OFB)

```python
iv = Crypto.Random.get_random_bytes(16)


cipher = Crypto.Cipher.AES.new(key, Crypto.Cipher.AES.MODE_OFB, iv=iv)


print(f'vector de inicialización: {b64encode(iv)}')


print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))
print(cipher.encrypt(b'Atacad a las dos'))

c = cipher.encrypt(b'Atacad a las dos')
```

# Receptor (modo OFB)

```python
decipher = Crypto.Cipher.AES.new(key, Crypto.Cipher.AES.MODE_OFB, iv=iv)

print(decipher.decrypt(c))
```

```python
print(decipher.decrypt(c))
print(decipher.decrypt(c))
print(decipher.decrypt(c))
print(decipher.decrypt(c))
print(decipher.decrypt(c))
```
